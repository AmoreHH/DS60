```python
#step2: 如果有關連性，運用 KNN ，取出 Age 、 Pclass、 Sex的資料，以 Sex 與 Pclass 補 Age 遺失值。
import sklearn
```

```python
df=df_train[['Age', 'Pclass', 'Sex']] 
```

```python
df #sex要做轉換
```

```python
#透過 preprocessing 轉換

from sklearn import preprocessing
#下面兩行程式，讓 Male=1, FeMale-0.

le = preprocessing.LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
print(df['Sex'])
```

```python
df
```

```python
from sklearn.metrics.pairwise import nan_euclidean_distances
display(pd.DataFrame(nan_euclidean_distances(df)))
```

```python
# KNN
# 設定 k 值
value_neighbors=1
from sklearn.impute import KNNImputer
#初始化: Initialize KNNImputer
imputer = KNNImputer(n_neighbors=value_neighbors)
df_filled = pd.DataFrame(imputer.fit_transform(df))
#display(data[missing_index.values==True])
#display(df_filled[missing_index.values==True])
display(df_filled)
```
